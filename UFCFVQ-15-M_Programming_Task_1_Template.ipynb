### Requirement FR1 - Develop a function to find the arithmetic mean

```python
# add code here
# The function mean(x) returns the arithmetic mean rounded to 2 decimal places
def mean(x): # x is a list
    n = len(x)
    mean = round(sum(x)/n,2)
    return mean
```

### Requirement FR2 - Develop a function to find the standard deviation

```python
# add code here

# using formula for population standard deviation
def stan_dev(x):  # x is a list
    #sutract the mean of x fom every item in the list and find the square
    #put the result in an empty list c
    #add all items in list c, divide by the population size N, and return the square root to 2 decimal places
    c = [] 
    for a in x:
        N = len(x)  #size of population,N
        b = (a-mean(x))**2
        c.append(b)
    stan_dev = round((sum(c)/N)**(1/2),2)
    return stan_dev
```

```python
x = [29, 17, 28, 6, 14, 7, 4, 27, 21, 15, 10, 16, 24, 26, 3, 11, 13, 8, 23, 9, 0, 22, 12, 2, 18, 19, 5, 1, 20, 25]
stan_dev(x)
```

### Requirement FR3 - Develop a function to find the median 

```python
# add code here

def median(x): # x is a list
    #first sort the list
    #check if list length is even or odd and return the element at the middle index number
    x.sort()
    for i in x:
        if len(x)%2 == 1:
            i = int((len(x)+1)/2 - 1)
            return x[i]
        else:
            i1 = int(len(x)/2 - 1)
            i2 = int(i1 + 1)
            return (x[i1] + x[i2])/2
           
```

### Requirement FR4 - Develop a function to find the skewness

```python
# add code here
def skewness(x):
    #sutract the mean of x fom every item in the list and find the cube
    #put the result in an empty list c
    #add all items in list c, divide by the population size N minus 1, and multiply by the cube of the standard deviation
    
    c = []
    for a in x:
        N = len(x)
        b = (a-mean(x))**3
        c.append(b)
    skewness = round(sum(c)/((N - 1)*stan_dev(x)**3),4) #rounded to 4 decimal places
    return skewness
```

### Requirement FR5 - Develop a function to read a single column from a CSV file

```python
# add code here

def single_col(csv,col_num):
    with open(csv,'r') as file:
        csv_edit1 = file.read().splitlines()
        i = 0
        col_values = []
        try: #catch and advice user on UnboundLocalError if user provides column number outside column range
        #split the items in csv_edit1 based on the comma separator
        #separate the column names from the column values
            for item in csv_edit1:
                csv_edit2 = item.split(',')
                if i < 1 and col_num < len(csv_edit2):
                    col_name = csv_edit2[col_num]     
                elif 0<i<len(csv_edit1) and col_num < len(csv_edit2):
                    col_values.append(csv_edit2[col_num]) 
                i+=1
        
            col_values = list(map(float,col_values)) #convert to float to achieve specification and enable further processing
            return col_name,col_values
        except UnboundLocalError:
            
            print("Error! Column number must be less than",len(csv_edit2))
```

### Requirement FR7 - Develop a function to generate a set of statistics for a given data file

```python
def csv_stats(csv,col_num):
    with open(csv,'r') as file:
        #split each item in the file in based on lines and then the comma separator
        edited_csv1 = file.read().splitlines()
        for item in edited_csv1:
            edited_csv2 = item.split(',')
        col_num = 0
        full_list = []
        #use the FR5 function above to get each column and add to a full list of columns
        try: #catch and advice user on any error if user provides column number outside column range
            while col_num < len(edited_csv2):
                each_column = single_col(csv,col_num)
                col_num+=1
                full_list.append(each_column)
            col_num = 0
            col_name = ["Stats"]
            col_values_lists = []
            full_stat_set = []
            while col_num < len(edited_csv2):
                for items in full_list:
                    #separate column names and lists of column values into different lists
                    col_name.append(items[0])
                    col_values_lists.append(items[1])

                for element in col_values_lists:
                    #generate set of statistics for each list of column values
                    #put results in a list containing full set of statistics for all columns
                    x = col_values_lists[col_num]
                    stat_set = [mean(x),stan_dev(x),median(x),skewness(x)]
                    full_stat_set.append(stat_set)
                    col_num+=1
            #include statistic description for clarity
            #zip the column names and statistics values lists in a dictionary
            stat_values = [["Mean","Stdev","Median","Skewness"]]
            stat_values.extend(full_stat_set)
            stat_dict = dict(zip(col_name,stat_values))  
            return stat_dict
        except:
            
            print("Error! Column number must be less than",len(edited_csv2))

    
```
